# Método da Bisseção e Método de Simpson

Neste trabalho vamos utilizar o método da bisseção que é um algoritmo de procura de raízes que divide um intervalo
ao meio e seleciona o semi-intervalo no qual uma mudança de sinal ocorre. Continua este processo até que o erro 
seja inferior ou igual ao erro definido. 
Uma vez definidos os valores dos zeros da função h(x), vamos utilizar o método de Simpson, uma técnica de integração numérica usada para encontrar a área aproximada entre duas curvas.

No código que se encontra em baixo, utilizámos as duas funções atribuídas pelo professor f(x) e g(x).

```python
import numpy as np
import matplotlib.pyplot as plt
from scipy.optimize import fsolve
def f(x):
    return np.sqrt(x) + 0.2

def g(x):
    return 2 * np.sin(x)

def h(x):
    return np.sqrt(x) + 0.2 - 2 * np.sin(x)

x = np.linspace(0, 7, 100)

# Calculando os valores das funções para cada valor de x
y1 = f(x)
y2 = g(x)
y3 = h(x)
# Plotando as funções
plt.plot(x, y1, label='f(x)')
plt.plot(x, y2, label='g(x)')
plt.plot(x, y3, label='h(x)')
# Adicionando rótulos aos eixos e uma legenda
plt.xlabel('x')
plt.ylabel('y')
plt.legend()

# Exibindo o gráfico
plt.show()

# Método da Bisseção
def bissecao(f, A, B, TOL, NMAX):
    i = 0
    while i < NMAX:
        P = (A + B) / 2.0
        if f(P) == 0 or (B - A) / 2 < TOL:
            print('iter=', i + 1, 'Aprox=', P)
            print('Erro=', (B - A) / 2)
            return P
        else:
            if f(A) * f(P) > 0:
                A = P
            else:
                B = P
            print('iter=', i + 1, 'Aprox=', P)
            print('Erro=', (B - A) / 2)
        i += 1
    return P

a = bissecao(h, 0, 1, 1e-2, 100)
print(a)

b = bissecao(h, 2, 3, 1e-2, 100)
print(b)

# Método de Simpson
def simpson(func, a, b):
    h = (b - a) / 2
    x = a + h
    integral = (h / 3) * (func(a) + 4 * func(x) + func(b))
    return integral

integral = simpson(h, a, b)
print(integral)
# Vamos agora calcular a área mas com os valores exatos para podermos comparar os dois resultados
# Encontrando as raízes de h(x) no intervalo escolhido
initial_guess_1 = 1  
initial_guess_2 = 3  

root1 = fsolve(h, initial_guess_1)
root2 = fsolve(h, initial_guess_2)

print("As raízes de h(x) no intervalo [0,4] são:")
print("Raiz 1:", root1)
print("Raiz 2:", root2)
# Vamos calcular agora a área,o integral pelo método de simpson, do intervalo correto
integral_certo=simpson(h, root1, root2)
print("O valor do integral_certo é:", integral_certo)
```
